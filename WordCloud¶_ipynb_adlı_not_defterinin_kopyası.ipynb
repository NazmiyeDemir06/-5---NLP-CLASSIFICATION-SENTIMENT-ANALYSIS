## 5.WordCloud

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
df = pd.read_csv('NewData2.csv')
```

```python
df.head()
```

```python
df.text[0]
```

#### 5.1. WordCloud (5 Star) - 4226 reviews

```python
star_mask = np.array(Image.open("wc.png"))
# 5 yldzl incelemeler iin kelime bulutu oluturma
wordcloud = WordCloud(width=800, height=400,
                      background_color='white',
                      mask=star_mask,
                      contour_color='black',
                      contour_width=1).generate(' '.join(df[df['stars'] == 5]['text']))

# Kelime bulutunu grselletirme
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('5 Yldzl ncelemeler iin Kelime Bulutu')
plt.show()
```

#### 5.2. WordCloud (1 Star) - 1051 reviews

```python
# 1 yldzl incelemeler iin kelime bulutu oluturma
wordcloud_one_star = WordCloud(width=800, height=400, background_color='white').generate(' '.join(df[df['stars'] == 1]['text']))

# Kelime bulutunu grselletirme
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_one_star, interpolation='bilinear')
plt.axis('off')
plt.title('1 Yldzl ncelemeler iin Kelime Bulutu')
plt.show()
```

```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image

# Yldz eklinde bir maske oluturma
star_mask = np.array(Image.open('cloud.png'))  # Burada 'star_shape.png' dosyasnn yolunu belirtin

# 1 yldzl incelemeler iin kelime bulutu oluturma
wordcloud_one_star = WordCloud(width=800, height=400,
                                background_color='white',
                                mask=star_mask,
                                contour_color='black',
                                contour_width=1).generate(' '.join(df[df['stars'] == 1]['text']))

# Kelime bulutunu grselletirme
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_one_star, interpolation='bilinear')
plt.axis('off')
plt.title('1 Yldzl ncelemeler iin Kelime Bulutu')
plt.show()
```

#### 5.3. WordCloud (Positive) - Mean: 4.01 - 9167 reviews

```python
positive_reviews = df[df['stars'] >= 4]

# Ortalama puan hesaplama ve kelime bulutunu oluturma
mean_rating = positive_reviews['stars'].mean()
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(positive_reviews['text']))

# Kelime bulutunu grselletirme
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title(f'Kelime Bulutu - Ortalama: {mean_rating:.2f} - {len(positive_reviews)} nceleme')
plt.show()
```

#### 5.4.WordCloud (Negative) - Mean: 2.03 - 1167 reviews

```python
negative_reviews = df[df['stars'] <= 3]

# Ortalama puan hesaplama ve kelime bulutunu oluturma
mean_rating = negative_reviews['stars'].mean()
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(negative_reviews['text']))

# Kelime bulutunu grselletirme
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title(f'Kelime Bulutu - Ortalama: {mean_rating:.2f} - {len(negative_reviews)} nceleme')
plt.show()
```
